```python
## 코로나 데이터를 로드
## 상위 폴더로 이동 -> csv폴더로 이동 -> corona.csv 파일 지정
import pandas as pd

corona = pd.read_csv('../csv/corona.csv')
corona.head()
```

```python
## 컬럼을 삭제, drop()함수를 이용
## axis라는 매개변수의 값을 기준으로 행, 열을 지정
## inplace 매개변수의 값을 기준으로 기준이 되는 데이터를 변경을 할지 지정

corona.drop('Unnamed: 0', axis=1, inplace=True)
# corona.drop('Unnamed: 0', axis=1)
```

```python
corona.head()
```

```python
## 컬럼의 이름을 변경
## 컬럼의 목록 출력 방법

corona.columns
corona.columns = [
    "등록일시","총 사망자", "총 확진자", "게시글 번호", "기준시간", 
    "기준일", "수정일시", "누적 의심자", "누적 확진률"
    ]
```

```python
corona.columns
```

```python
corona.head()
```

```python
## 게시글 번호, 기준일, 기준시간 컬름을 삭제
#corona.drop(['게시글 번호','기준시간','기준일'], axis=1)
corona.drop(['게시글 번호','기준시간','기준일'], axis=1, inplace=True)
```

```python
corona.head()
```

```python
## 데이터프레임의 정렬을 변경
## 등록일시를 기준으로 오름차순 정렬
## sort_values([기준이 되는 컬럼명])

corona.sort_values(['등록일시'])
```

```python
## asceding 매개변수의 값이 True(기본값)이면 오름차순, Fasle면 내림차순

corona.sort_values(['등록일시'], ascending=0)
```

```python
## 데이터프레임의 정렬을 변경
## 등록일시를 기준으로 오름차순 정렬
## sort_values([기준이 되는 컬럼명])

corona.sort_values(['등록일시'], inplace=True)
```

```python
corona.head()
```

```python
##인덱스의 값을 초기화
## reset_index()
corona.reset_index(drop=True, inplace=True)
```

```python
## 결측치 조회
corona.info()
```

```python
corona.isna().sum()
```

```python
## 데이터 간의 차이를 출력하는 함수
##[1번째 방법]  diff(n) : n은 인덱스의 이동 회수( n은 기본값 1)
corona['일일 확진자'] = corona['총 확진자'].diff()
```

```python
corona.head()
```

```python
corona['일일 사망자'] = corona['총 사망자'].diff()
```

```python
corona.head()
```

```python
## 데이터 간의 차이를 출력하는 함수
## [2번째 방법] shift
corona['총 사망자'] - corona['총 사망자'].shift(1)
```

```python
## 결측치를 다른 값으로 대체
## fillna(n) : 결측치를 n으로 채워준다.
##일일 확진자와 일일 사망자의 결측치를 0으로 대체한다. 
corona[['일일 확진자','일일 사망자']] = corona[['일일 확진자','일일 사망자']].fillna('-')
```

```python
corona.head()
```

```python
## fillna() method 매개변수 -> ffill(바로 전의 data로 대체), bfill(바로 뒤의 data로 대체)
corona.iloc[690:695]
```

```python
corona['누적 의심자'] = corona['누적 의심자'].fillna(method= 'ffill')
```

```python
corona.iloc[690:695]
```

```python
x = [1,2,3,4]
y1 = [12,30,60,70]
y2 = [90,80,30,20]
##한글이 깨지는 현상을 막기위해 폰트 변경
import platform

if platform.system() == 'Darwin' : ## Mac인 경우
    plt.rc('font', family ='AppleGotic')
else:
    plt.rc('font', family = 'Malgun Gothic')    

##선의 색
plt.plot(x,y1,'r')
##선의 종류
plt.plot(x,y2,'--')
## x축의 이름
plt.xlabel('Class')
## y축의 이름
plt.ylabel('Count')
## tilte
plt.title('그래프')
plt.show()
```

```python
## subplot(행의 수, 열의 수, 위치)
## 그래프 2개 그린다는 가정
## subplot을 이용해서 위치 지정
plt.subplot(2,1,1)
plt.plot(x,y1,'r')
plt.subplot(2,1,2)
plt.plot(x,y2,'y')
plt.show()
```
