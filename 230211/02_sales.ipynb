```python
## 예제 문제
1. csv 폴더에 있는 sales records.csv 파일 로드
2. ' Order ID', 'Total Revenue', 'Total Cost', 'Total Profit' 컬럼을 삭제
3. 컬럼의 이름을 변경 [ '대륙', '국가', '타입', '채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
4. 대륙, 국가를 기준으로 오름차순 정렬
5. 인덱스를 초기화 (기존의 인덱스는 삭제)
```

```python
import pandas as pd

## 파일 불러오기(상대주소: 상위 폴더 -> 하위 폴더 -> 파일 지정)
sales = pd.read_csv('../csv/Sales records.csv') 
sales.head(1)
```

```python
## 불필요 컬럼(열) 삭제하고 저장하기
sales.drop(['Order ID','Total Revenue', 'Total Cost', 'Total Profit'], axis =1, inplace = True)
sales.head(1)
```

```python
## 컬럼 이름 변경
sales.columns = ['대륙', '국가', '타입', '채널', '우선순위', '주문일자', '발송일자', '판매단위', '단가', '원가']
sales.head(1)
```

```python
sales.sort_values(['대륙','국가'], ascending=True, inplace = True)  ## 오름차순으로 정렬
```

```python
sales.head()
```

```python
sales.head()
```

```python
## 인덱스(번호)를 초기화(기존의 인덱스는 삭제)
sales.reset_index(drop = True, inplace = True)
```

```python
sales.head()
```

```python
## 총 단가, 총 원가, 총 이윤 컬럼을 생성
## 총 단가 = 판매단위 * 단가
sales['총 단가'] = sales['판매단위'] * sales['단가']
## 총 원가 = 판매단위 * 원가
sales['총 원가'] = sales['판매단위'] * sales['원가']
## 총 이윤 = 총 단가 - 총 원가
sales['총 이윤'] = sales['총 단가'] - sales['총 원가']
## 총 이윤 다른 방법
## sales['총 이윤'] = sale['판매단위']*(sales['단가'] - sales['원가'])
```

### 그룹화
- groupbby(그룹화 기준이 되는 컬럼)

```python
sales.groupby('국가')
```

```python
# 국가별 총 합계
sales.groupby('국가').sum()
```

```python
## 국가별 총 이윤 확인
sales.groupby('국가').sum()['총 이윤'] 
```

```python
# 리스트 형태로 출력
sales.groupby('국가').sum()[['총 이윤']]  
```

```python
## 가장 많은 이윤을 내는 대륙은 어디인가? 
## 그룹화 - 합계 - 내림차순
```

```python
sales.groupby('대륙').sum().sort_values(['총 이윤'], ascending=False)
```

```python
sales_C = sales.groupby('대륙').sum()[['총 이윤']]
sales_C
```

```python
sales_pro_sort = sales_C.sort_values('총 이윤', ascending= False)
sales_pro_sort 
```

```python
## 국가별 총 이윤이 높은 상위 10개의 국가를 출력
test2 = sales.groupby('국가').sum()[['총 이윤']]
test2
```

```python
## 총 이윤을 기준으로 내림차순
group_df = test2.sort_values('총 이윤', ascending=False)
```

```python
## 상위 10개
top10 = group_df.head(10)
top10
```

```python
#하위 10개
bottom10 = group_df.tail(10)
bottom10
```
