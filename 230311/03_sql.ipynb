```python
#라이브러리 로드
import pymysql
import pandas as pd
```

```python
#서버 정보와 서버의 접속
_db = pymysql.connect(
    host = 'localhost',
    user = 'root',
    password = '123456',
    db = 'test',
    port = 3306, #기본 port번호 
    
)
```

```python
# cursor 생성
# sql 쿼리문을 실행시키기 위해 만들어야 하는 필수 변수 
# DictCursor : 쿼리문이 실행되고 나온 결과물을 딕셔너리형태로 받아온다. 
cursor = _db.cursor(pymysql.cursors.DictCursor)
```

```python
# '''  장문의 문자열을 사용할때 칸을 바꾸어도 가능   ''''
sql = '''
    select 
    *
    from
    corona
'''
cursor.execute(sql)    # execute는 보내서 실행만 시킨다.

result = cursor.fetchall()  # fetchall는 실행한 결과를 받아온다.
print(result)

```

```python
pd.DataFrame(result).head()   # 데이터베이스의 정보를 데이터프레임화 한다.
```

```python
# insert 문
sql = '''
    insert into
    user
    values(
    'test1','1234','test1',0
    )
'''
cursor.execute(sql)  # 실행은 했지만, 아직 최종 변경되지는 않았다.
```

```python
_db.commit()  # 최종 데이터베이스에 변경 적용
```

```python
_id = input('id를 입력하세요')
_pw = input('password를 입력하세요')
_name = input('name을 입력하세요')
_gender = int(input('성별을 입력하세요: 남자 0 or 여자 1'))

sql2 = ''' insert into user values('{_id}','{_pw}','{_name}',{_gender}) '''
cursor.execute(sql2)
_db.commit()
```

```python
_id = input('id를 입력하세요')
_pw = input('password를 입력하세요')
_name = input('name을 입력하세요')
_gender = int(input('성별을 입력하세요: 남자 0 or 여자 1'))

sql3 = ''' insert into user values(%s, %s, %s, %s) '''
cursor.execute(sql3, [_id, _pw, _name, _gender])
_db.commit()
```
